### Cargamos las librerias necesarias

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Cargamos los datos

```python
url  = "https://raw.githubusercontent.com/MinCiencia/Datos-COVID19/master/output/producto23/PacientesCriticos.csv"
data = pd.read_csv(url)
data
```

## Pacientes criticos por fecha

```python
fecha = data.columns[1:].tolist()
p_criticos = data[fecha].values.tolist()[0]
```

```python
def labelonpoints(X, Y):
    n = len(X)
    for i in range (n):
        number = '{:,}'.format(int(Y[i])).replace(',', '.')
        plt.annotate(
            number, xy=(X[i], Y[i]),
            xytext =(0, 10),
            textcoords="offset points",
            ha ='center', va='bottom')

plt.figure(figsize=(20, 10), dpi=80)
plt.plot(fecha, p_criticos, 'o-')
plt.xticks(fecha, rotation='vertical')
labelonpoints(fecha, p_criticos)
plt.grid()
plt.show()
```
