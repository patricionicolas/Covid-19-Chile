```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact 
```

## Cargamos la data

```python
url = "https://raw.githubusercontent.com/MinCiencia/Datos-COVID19/master/output/producto1/Covid-19.csv"
data = pd.read_csv(url)
data.fillna(0)
data
```

## Curva de crecimiento de casos por comuna

```python
comunas = data["Comuna"].unique().tolist()
fechas  = data.columns[(data.columns != "Region") & (data.columns != "Codigo region") & (data.columns != "Comuna") & (data.columns != "Codigo comuna") & (data.columns != "Poblacion") & (data.columns != "Tasa")]
```

```python
def anotate(X,Y):
    for i in range(len(X)):
        number = '{:,}'.format(int(Y[i])).replace(',', '.')
        plt.annotate(
            number, xy=(X[i],Y[i]),
            xytext =(0, 10),
            textcoords="offset points",
            ha ='center', va='bottom')

def Create_graphic_for_comuna(comuna):
    data_Comuna = data[data["Comuna"] == comuna]  #Filtramos los datos por la comuna ingresada
    X           = fechas                          #Eje X correspondera a las fechas
    Y           = data_Comuna[fechas].values      #Eje Y correspondera a la cantidad de contagiados
    
    print (comuna)
    plt.figure(figsize=(15,10))
    plt.title("Crecimiento de los casos - " + comuna)
    plt.xticks(rotation=90)
    plt.grid()
    plt.plot(X, Y[0], "bo-")
    anotate(X, Y[0])
    plt.show()
    plt.clf()
```

```python
for comuna in comunas:
    Create_graphic_for_comuna(comuna)
```

```python
interact(Create_graphic_for_comuna,
         comuna = comunas)
None
```
